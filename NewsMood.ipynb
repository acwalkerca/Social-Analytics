```python
#Dependencies
import tweepy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


#Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()


#Twitter API Keys
from config import (consumer_key,
                    consumer_secret,
                    access_token,
                    access_token_secret)

# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
#Target Search Terms
target_user = ("@BBC", "@CBS", "@cnn", "@FoxNews", "@nytimes" )

#List of dictionaries for results
results_list = []

# Create variable for holding the oldest tweet
oldest_tweet = None

# Counter to keep track of the number of tweets retrieved

#Loop through each user
for user in target_user:
    
    #Variables for holding sentiments
    compound_list = []
    positive_list = []
    neutral_list = []
    negative_list = []
    timestamp = []
    tweet_number = []
    
    #Loop through all tweets
    for x in range(1):
        
        #Get all tweets from home feed
        public_tweets = api.user_timeline(user, count=100)
        
        counter = 1
        
        for tweet in public_tweets:
            
            results = analyzer.polarity_scores(tweet["text"])
            compound = results["compound"]
            pos = results["pos"]
            neu = results["neu"]
            neg = results["neg"]
            time = tweet["created_at"]
            tweet_count = counter
            
            compound_list.append(compound)
            positive_list.append(pos)
            neutral_list.append(neu)
            negative_list.append(neg)
            timestamp.append(time)
            tweet_number.append(tweet_count)
            
            counter +=1   
                    
    user_results = {
        "Username" : user,
        "Tweet Polarity" : compound_list,
        "Timestamp" : timestamp, 
        "Tweets Ago" : tweet_number}
    results_list.append(user_results)
```

```python
bbc_df = pd.DataFrame(results_list[0])
cbs_df = pd.DataFrame(results_list[1])
cnn_df = pd.DataFrame(results_list[2])
fox_df = pd.DataFrame(results_list[3])
nyt_df = pd.DataFrame(results_list[4])
```

```python
bbc = bbc_df.plot(kind = 'scatter', x = "Tweets Ago", y = "Tweet Polarity", color = "powderblue", marker = "o", edgecolors = "black", s = 60, label = "BBC", alpha = .75, zorder = 2);
cbs = cbs_df.plot(kind = 'scatter', x = "Tweets Ago", y = "Tweet Polarity", color = "green", marker = "o", edgecolors = "black", s = 60, label = "CBS",alpha = .75, zorder = 2, ax=bbc);
cnn = cnn_df.plot(kind = 'scatter', x = "Tweets Ago", y = "Tweet Polarity", color = "red", marker = "o", edgecolors = "black", s = 60, label = "CNN", alpha = .75, zorder = 2, ax=cbs);
fox = fox_df.plot(kind = 'scatter', x = "Tweets Ago", y = "Tweet Polarity", color = "darkblue", marker = "o", edgecolors = "black", s = 60, label = "Fox", alpha = .75, zorder = 2, ax=cnn);
nyt = nyt_df.plot(kind = 'scatter', x = "Tweets Ago", y = "Tweet Polarity", color = "yellow", marker = "o", edgecolors = "black", s = 60, label = "New York Times", alpha = .75, ax=fox, zorder = 2)
bbc.legend(title = "Media Sources", loc= 'best', bbox_to_anchor=(1, 1))
plt.grid(color = 'white')
nyt.set_facecolor('lavender')
x_vals = bbc_df["Tweets Ago"]
plt.xlim([x_vals.max()+5,x_vals.min()-5])
plt.title("Sentiment Analysis of Media Tweets (7/7/18)")
nyt.locator_params(axis='y', nbins=5)
```
